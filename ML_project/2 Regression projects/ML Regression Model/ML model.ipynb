```python
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pan
```

```python
data = pan.read_csv('height_and_weight.csv')
```

```python
data.drop(columns = ['Index'],inplace = True)
```

```python
# If we give x values as
# X = data['Weight(lbs)'].values
# then all values will be given to a single array, we need to avoid it
```

```python
H = data.drop(columns=['Weight(lbs)'])
# It will provide each X Key i.e. input as an array [[1],[2],...........200 times]
W = data.drop(columns=['Height(In)'])

# Training set  180 samples
X_inp = H.values[:180]
y_out = W.values[:180]

# testing set 
test_X = H.values[180:]
test_y = W.values[180:]
# len(test_X)
```

```python
# # AS we passed 180 sets to train and 20 sets to test it, we performed manual slicing
# # Same using scikitlearn
# from sklearn.model_selection import train_test_split
# X_inp,test_X,y_out,test_y = train_test_split(H,W,test_size=0.1)
# # Here 0.1 denotes 10% is for testing set
```

```python
# X_inp.values
```

```python
# data.plot.bar(subplots=True)
# data.head(50).plot.bar()
# data.head(50).plot.bar(subplots=True)
# X_inp.head(50).plot.bar()
```

```python
Model = LinearRegression()
Model.fit(X_inp,y_out)
Model
```

```python
# type(LinearRegression())
```

```python
Hypothesis = Model.predict(test_X)
```

```python
compare = pan.DataFrame({"Hypothesis":Hypothesis.flatten(),"y":test_y.flatten()})
```

```python
compare.plot()
```

```python
compare.plot.bar(figsize=(10,6))
```

```python
from matplotlib import pyplot
pyplot.scatter(test_X,test_y)
pyplot.plot(test_X,Hypothesis,'r')
```

```python
from sklearn import metrics
metrics.max_error(test_y,Hypothesis)
```
