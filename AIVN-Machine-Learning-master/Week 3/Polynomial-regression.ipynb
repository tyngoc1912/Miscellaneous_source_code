### 1. Polynomial Regression

Polynomial Regression (hồi quy đa thức) là dạng bài toán hồi quy trong đó biến $x$ phụ thuộc vào biến $y$ theo một đa thức bậc $n$. Ví dụ hàm giả thuyết: $h_\theta(x) = \theta_0 + \theta_1x + \theta_2x^2$ hoặc $h_\theta(x) = \theta_0 + \theta_1x + \theta_2\sqrt{x}$ ...

<img src="https://raw.githubusercontent.com/pickus91/Polynomial-Regression-From-Scratch/master/Figures/hypothesis.PNG" style="width:40%;height:40%;">


Mọi lý thuyết của bài toán hồi quy tuyến tính đều giống với bài toán hồi quy đa thức. Chỉ khác bây giờ hàm giả thuyết không còn là đường thẳng mà là hàm đa thức bậc $n$.

### 2. Code Demo 

Trong bài toán này mang tính giới thiệu nên chúng ta sẽ sử dụng thư viện __sklearn__ để tìm nghiệm của bài toán.

#### 2.1 Import Thư Viện 

- pandas - thư việc dùng để load và hiển thị dữ liệu
- numpy - thư viện numpy 
- matplotlib - thư viện matplotlib dùng để trực quan hoá dữ liệu

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # Format này nhằm đưa giá trị về Matrix
y = dataset.iloc[:, 2].values
```

#### 2.2 Trực quan hoá dữ liệu

```python
plt.scatter(X, y, color = 'red')
plt.title('Position Level vs Salary')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

#### 2.3 Sử dụng Sklearn 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# degree: số bậc đa thức
# Đa thức bậc 4
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
# Kết quả của hồi quy đa thức
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Position Level vs Salary')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising kết quả Polynomial Regression 
X_grid = np.arange(min(X), max(X), 0.1)     # Đặt kích thước nhỏ cho đường cong mượt 
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Position Level vs Salary')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

#### 2.4 Dự đoán kết quả

```python
# Predicting a new result with Polynomial Regression
predict = lin_reg.predict(poly_reg.fit_transform(np.array([[6.5]])))
print('Kết quả dự đoán:', predict[0])
```
